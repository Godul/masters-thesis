# Single linear regression

We have not many samples (6), so `statsmodels` will give us warnings. We turn them off.

```python
import warnings
warnings.filterwarnings('ignore')
```

## Redis YCSB 6x10

```python
from helpers import (
    get_means,
    fit_regression,
    draw_regression_graph,
)

experiment_name = 'redis_ycsb_6x10'

exp_name, df = next(get_means(experiment_name))
results = fit_regression(data=df, formula='app_latency ~ instances_n')
print(results.summary())
print(results.params)

for exp_name, df in get_means(experiment_name):
    results = fit_regression(data=df, formula='app_latency ~ instances_n')
    draw_regression_graph(results, df, metric_name='app_latency', variable='instances_n')
```
