# Plot OS metrics

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

from helpers.load_data import (
    get_experiments_paths,
    get_data_with_metrics,
)


def plot_os_metric(df: pd.DataFrame, metric: str):
    try:
        df.plot(y=metric, x='time', title=metric)
    except ValueError as e:
        print(f'Cannot plot {metric}.')
        print(str(e))
        print(f'Value: {df[metric][0]}')

        
def plot_all_os_metrics(df: pd.DataFrame):
    for column_name in df.columns:
        plot_os_metric(df, column_name)


experiments_path = '../../data/basic'
```

## Redis YCSB vs Redis YCSB

```python
experiment_name='redis_ycsb_1_redis_ycsb_11'
instances_n=12
path = get_experiments_paths(experiment_name, experiments_path)[0]
df = get_data_with_metrics(path, instances_n=instances_n, cpu_window=30)
df['time'] = pd.to_datetime(df['cbtool_time'], unit='s')
```

```python
plot_os_metric(df, 'app_latency')
```

### node_memory_Committed_AS_bytes

Memory information field Committed_AS_bytes

It means how much total memory space all applications have reserved for use. It does not necessarily mean how much they are actually using.


```python
plot_os_metric(df, 'node_memory_Committed_AS_bytes')
```

### node_memory_MemFree_bytes

Memory information field MemFree_bytes

```python
plot_os_metric(df, 'node_memory_MemFree_bytes')
```

### node_memory_SUnreclaim_bytes

Reclaimable SLAB memory can be reused by the kernel for other things, unreclaimable can't. Where any given SLAB allocation ends up accounted depends on the properties of the pool it's allocated in, which in turn means it's a property of the allocation itself.

Slightly OT, but for what it's worth, that huge chunk of unreclaimable SLAB memory is probably the ZFS ARC.

```python
plot_os_metric(df, 'node_memory_SUnreclaim_bytes')
```

### node_filesystem_free_bytes_device__dev_mapper_ubuntu__vg_ubuntu__lvfstype_ext4mountpoint___

Filesystem free space in bytes.

The /dev/mapper/ubuntu--vg-root in place of a traditional /dev/sdxN block device just indicates that you chose to install the system using LVM2 logical volume management.

```python
plot_os_metric(df, 'node_filesystem_free_bytes_device__dev_mapper_ubuntu__vg_ubuntu__lvfstype_ext4mountpoint___')
```

### node_cpu_seconds_total_cpu_6mode_iowait_



```python
plot_os_metric(df, 'node_cpu_seconds_total_cpu_6mode_iowait_')
```

### node_cpu_seconds_total_cpu_18mode_iowait_

iowait is time that the processor/processors are waiting (i.e. is in an idle state and does nothing), during which there in fact was outstanding disk I/O requests. This usually means that the block devices (i.e. physical disks, not memory) is too slow, or simply saturated.

```python
plot_os_metric(df, 'node_cpu_seconds_total_cpu_18mode_iowait_')
```
