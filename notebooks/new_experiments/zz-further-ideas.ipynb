# Further ideas

### Eksperymenty

* **Powtórzyć eksperymenty na maszynie z >6 rdzeniami.**

* Sprawdzić jak radzi sobie model przy większej liczbie typów tasków (do tej pory mamy eksperymenty tylko dla 2).

* Sprawdzić jak dokładane instancje wpływają na resztę instancji, a nie tylko na 1. Czy kolejnośc dokładnania instancji ma wpływ?


* Przeprowadzić eksperymenty na różnych maszynach i zbadać współczynniki inferencji.
    * Czy współczynniki są takie same? (Hipoteza: nie)
    * Problem: "workloady na czas" (np. sysbench) vs "workloady na zadanie" (np. linpack) i ich rozmiar


### Predykcje

* Przewidywać wielkość współczynnika interferencji na podstawie zużycia zasobów.
    * zmienne objaśniające: os_metrics
    * zmienna objaśniana: $\alpha_{newtype, linpack}$ lub $\alpha_{linpack, newtype}$
    * dane: eksperymenty i pozyskane z nich os_metrics i $\alpha_{type, linpack}$ lub $\alpha_{linpack, type}$


* Przewidywać wielkość współczynnika interferencji na podstawie wyników eksperymentów z częścią typów.
    * model: ($\alpha_{newtype, linpack}$, $\alpha_{newtype, redis\_ycsb}$, $\alpha_{newtype, hadoop}$) &#8594; $\alpha_{newtype, oldtype}$
    * dane: ($\alpha_{x, linpack}$, $\alpha_{x, redis\_ycsb}$, $\alpha_{x, hadoop}$, $\alpha_{x, oldtype}$)
    * problem: potrzeba więcej różnych workloadów i dużo więcej eksperymentów
    

### Klastrowanie

* Zrobić eksperymenty dla większej ilości workloadów i spróbować zaklasyfikować niektóre zadania jako jeden typ. Przy okazji trzeba będzie nauczyć się wyznaczać rozmiar zadania. Nowe workloady można pozyskać poprzez:
    * Zmianę `load_profile`, np. pagerank, terasort, wordcount itd. dla hadoopa.
    * Zmianę `load_level` (i innych parametrów), wpływając na intensywność znanych już nam eksperymentów.
    * Wypróbowanie nowych typów workloadów.


### Statystyka

* Porównać zużycie zasobów przez poszczególne zadania i zobaczyć, czy można za pomocą tego wytłumaczyć, które zadania mają na siebie największy wpływ.

* Przetestować statystyczne siły przewidywania modelu (na ile model tłumaczy więcej niż cpu load).
